# **Nettoyage Sous_nutrition**

```python
sous_nutrition = pd.read_csv('sous_nutrition.csv')
sous_nutrition.head()
```

```python
sous_nutrition.info()
```

```python
# Convertir les données en float / numerique
sous_nutrition['Valeur'] = pd.to_numeric(sous_nutrition['Valeur'], errors='raise')
sous_nutrition.fillna(0, inplace=True)
```

```python
# Rechercher l’élément à la position 60
sous_nutrition.iloc[60].Valeur
```

```python
# Convertir les données en numerique
sous_nutrition['sous_nut'] = pd.to_numeric(sous_nutrition['Valeur'], errors='coerce')
```

```python
# Regarder les données NaN
sous_nutrition[sous_nutrition.sous_nut.isnull()]
```

```python
# Convertir les données NaN en 0
sous_nutrition.fillna(0, inplace=True)
```

```python
# Normaliser à l'unité
sous_nutrition['sous_nut'] *= 1000000
```

```python
# Remplacer le nom Zone par Pays
sous_nutrition.rename(columns={"Zone": "Pays"}, inplace=True)
sous_nutrition.head()
```

```python
# Remplacer le nom Valeur par MoyHabSousAli
sous_nutrition.rename(columns={"Valeur": "MoyHabSousAli"}, inplace=True)
sous_nutrition.head()
```

```python
# Convertir le type de valeur MoyHabSousAli 'objet' par 'float'
sous_nutrition['MoyHabSousAli'] = pd.to_numeric(sous_nutrition['MoyHabSousAli'], errors='coerce')
sous_nutrition.fillna(0, inplace=True)
sous_nutrition.info()
```

```python
# Selectionner l’année 2017 via .query pour les intervalles
df_sous_nutrition_2017=sous_nutrition.query("Année =='2016-2018'")
print(df_sous_nutrition_2017)
```

```python
# Définir la DF "sous_nut_2017" et filtrer l’année 2017 puis garder les colonnes "Pays", "sous_nut"
sous_nut_2017 = df_sous_nutrition_2017[df_sous_nutrition_2017['Année'] == '2016-2018'][["Pays", "sous_nut"]]
```

# 🗺 **Disponibilité alimentaire mondiale**

```python
df_dispo_alim = pd.read_csv('dispo_alimentaire.csv')
df_dispo_alim.head()
```

```python
# Remplacer les NaN par 0 & convertir les données unités en milliers 
df_dispo_alim.fillna(0, inplace=True)

colonnes_millier_tonnes = ['Aliments pour animaux', 'Disponibilité intérieure', 'Exportations - Quantité',
                        'Importations - Quantité', 'Nourriture', 'Pertes', 'Production',
                        'Semences', 'Traitement', 'Variation de stock', 'Autres Utilisations']

for e in colonnes_millier_tonnes:
    df_dispo_alim[e] *= 1000000
```

```python
# Remplacer le nom Zone par Pays
df_dispo_alim.rename(columns={"Zone": "Pays"}, inplace=True)
df_dispo_alim.head()
```

```python
# Merger la table Population 2017 avec la table disponibilité alimentaire
dispo_alim_pop_2017 = pop_2017.merge(df_dispo_alim) 
dispo_alim_pop_2017.head()
```

```python
# Prendre le merge "dispo_alim_pop_2017" en selectionnant la colonne [Dispo alim...] et mutiplier par la dispo_alim_pop_2017 en selectionnant la colonne [Hab]
dispo_alim_pop_2017['dispo_kcal_pop']=(dispo_alim_pop_2017["Disponibilité alimentaire (Kcal/personne/jour)"]*dispo_alim_pop_2017["Hab"])
```

```python
# Disponibilité alimentaire total
df_sum_dispo = dispo_alim_pop_2017['dispo_kcal_pop'].sum()
```

```python
# Dispo alimentaire totale en kcal par année
f"dispo alimentaire totale en kcal par année: {df_sum_dispo*365}"
```

```python
# Dispo alimentaire totale en kcal par jour
f"dispo alimentaire totale en kcal par jour: {df_sum_dispo}"
```

# 👩👨🍖 **Calcul du nombre théorique de personnes qui pourraient être nourries** 
Femme = 2200 Kcal /jour & Homme = 2600 Kcal /jour = moyenne : 2400 Kcal/jour/personne

```python
# Disponibilité totale / par le nombre de calories moyens qu'il faut pour un humain par jour
sum_pers_nourries = df_sum_dispo/2400
print(sum_pers_nourries)
round(sum_pers_nourries)
```

```python
# Proportion de personne nourrie
proportion_pers_nourries = sum_pers_nourries*100/total_pop_2017
round(proportion_pers_nourries, 2)
```
